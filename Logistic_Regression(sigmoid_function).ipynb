```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
x = np.linspace(-10., 10., 100)

y = 1.0 / (1 + np.exp(-x))
```

```python
plt.plot(x, y, lw=3)
plt.show()
```

# Logistic Regression Cost function

```python
x = np.arange(0.001, 1., 0.001)
```

```python
plt.figure(figsize=(12, 4), dpi=150)

plt.subplot(1, 3, 1)
plt.plot(x, np.log(x))
plt.ylim(-5, 5)
plt.title('$\log (x)$')

plt.subplot(1, 3, 2)
plt.plot(x, -np.log(x))
plt.ylim(-5, 5)
plt.title('$-\log (x)$')

plt.subplot(1, 3, 3)
plt.plot(x, -np.log(1-x))
plt.ylim(-5, 5)
plt.title('$-\log(1-x)$')

plt.show()
```

```python
plt.figure(figsize=(6, 3), dpi=150)

plt.subplot(1, 2, 1)
plt.plot(x, -np.log(x))
plt.title('$-\log(x)$')

plt.subplot(1, 2, 2)
plt.plot(x, -np.log(1-x))
plt.title('$-\log(1-x)$')

plt.show()
```

```python
plt.figure(figsize=(4, 3), dpi=150)

plt.plot(x, -np.log(x), label='$y=1$')
plt.plot(x, -np.log(1-x), label='$y=0$')
plt.legend(loc=9)
plt.title('$-y\log(x)-(1-y)\log(1-x)$')

plt.show()
```
